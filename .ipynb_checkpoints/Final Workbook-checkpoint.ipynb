```python
from selenium import webdriver
import time
import chromedriver_binary
from bs4 import BeautifulSoup
from collections import defaultdict
import random
import pickle
import pandas as pd
import numpy as np
import json
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
```

# Composing dataframe
---

```python
# To load dateframe from pickle
with open('dateframe_pickle', 'rb') as file:
    df = pickle.load(file)
    
# To load scraped temperture
with open('temp_pickle', 'rb') as file:
    temp_col = pickle.load(file)
```

```python
df.occupancydatetime
```

```python
len(temp_col)
```

```python
# Combine parking availability and temperature historical to the same dataframe
for i in ['high', 'low', 'avg', 'precipitation']:
    df[i] = df['occupancydatetime'].map(lambda x: temp_col[x.strftime("%Y/%m/%d")][i])
```

```python
# Transform the main dataframe to a smaller one
small_df = df[['occupancydatetime', 'availableparkingspace', 'high', 'low', 'avg', 'precipitation']]

# Adding 'day of week' and 'hour' to our dataframe.
small_df['date'] = small_df.occupancydatetime.dt.strftime('%Y/%m/%d')
small_df['day'] = small_df['occupancydatetime'].dt.dayofweek
small_df['hr'] = small_df['occupancydatetime'].dt.hour
small_df['Q'] = small_df['occupancydatetime'].dt.quarter

# Reorder columns
small_df = small_df['date, Q, day, hr, availableparkingspace, high, low, avg, precipitation'.split(', ')]
```

```python
result_df = small_df.groupby(['date','day','hr'], as_index=False).mean()
```

# Splitting Train, Validate and Test
---

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, LassoCV, Ridge, RidgeCV
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.metrics import r2_score
```

```python
# Defining x, y
x, y = result_df.drop(['availableparkingspace', 'date'], axis=1), result_df['availableparkingspace']
print(x.shape)
x.head()
```

```python
# Spliting 20% as test data
x_trainval, x_test, y_trainval, y_test = train_test_split(x, y, test_size=.2, random_state=8)
print(len(x), len(x_test))
```

```python
x_train, x_val, y_train, y_val = train_test_split(x_trainval, y_trainval, test_size=.25, random_state=3)
print(len(x_train), len(x_test))
```

```python
x_train.shape
```

```python
x_val.shape
```

```python
#set up the 3 models we're choosing from:

lm = LinearRegression()

#Feature scaling for train, val, and test so that we can run our ridge model on each
scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train.values)
x_val_scaled = scaler.transform(x_val.values)
x_test_scaled = scaler.transform(x_test.values)

lm_reg = Ridge(alpha=1)

#Feature transforms for train, val, and test so that we can run our poly model on each
poly = PolynomialFeatures(degree=2) 

x_train_poly = poly.fit_transform(x_train.values)
x_val_poly = poly.transform(x_val.values)
x_test_poly = poly.transform(x_test.values)

lm_poly = LinearRegression()

#LASSO
# Run the cross validation, find the best alpha, refit the model on all the data with that alpha

alphavec = 10**np.linspace(-2,2,200)
lasso_model = LassoCV(alphas = alphavec, cv=5)
```

```python
import statsmodels.api as sm

from statsmodels.tools import add_constant

model = sm.OLS(y_train, add_constant(x_train), hasconst=True)

fit = model.fit()
fit.summary()
```

```python
#validate
lm.fit(x_train, y_train)
print(f'Linear Regression train R^2: {lm.score(x_train, y_train)}')
# print(f'Linear Regression train R^2: {lm.score(x_train, y_train):.3f}')

lm.fit(x_train, y_train)
print(f'Linear Regression val R^2: {lm.score(x_val, y_val)}')
# print(f'Linear Regression val R^2: {lm.score(x_val, y_val):.3f}')

lm_reg.fit(x_train_scaled, y_train)
print(f'Ridge Regression val R^2: {lm_reg.score(x_val_scaled, y_val)}')
# print(f'Ridge Regression val R^2: {lm_reg.score(x_val_scaled, y_val):.3f}')

lm_poly.fit(x_train_poly, y_train)
print(f'Degree 2 polynomial regression val R^2: {lm_poly.score(x_val_poly, y_val):.3f}')

lasso_model.fit(x_train_scaled, y_train)
print(f'LASSO val R^2: {lasso_model.score(x_val_scaled, y_val):.3f}')
```

# Result
---

### R Squared

```python
print(r2_score(y_test, lm.predict(x_test)))
print(r2_score(y_test, lm_reg.predict(x_test_scaled)))
print(r2_score(y_test, lm_poly.predict(x_test_poly)))
print(r2_score(y_test, lasso_model.predict(x_test_scaled)))
```

### Mean Squared Error

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(lm_poly.predict(x_val_poly), y_val)
```
